**Step 1: Importing the libraries**

```python
import numpy as np
import pandas as pd
```

**Step 2: Importing dataset**

```python
dataset = pd.read_csv("../../datasets/day-06-data.csv")

#All rows all columns except last (target)
X = dataset.iloc[:,:-1].values

#All rows but only one column (target)
Y = dataset.iloc[:,-1].values
```

**Step 3: Handling the missing data**

```python
from sklearn.impute import SimpleImputer

#Lets replace the missing values with the mean of column
imputer = SimpleImputer(missing_values = np.nan,
                        strategy = "mean")

#All rows to be considered for mean, and only column 1 and 2 as they have missing values
X[:,1:3] = imputer.fit_transform(X[:,1:3])
```

```python
#As you can see the NaN now replaced with mean of the column
print(X)
```

**Step 4: Encoding categorical data**

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    transformers=[
        ("encoder", OneHotEncoder(), [0])
    ],
    remainder="passthrough"
)

X = np.array(ct.fit_transform(X))
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder_Y = LabelEncoder()
Y = labelencoder_Y.fit_transform(Y)
```

**Step 5: Splitting the datasets into training sets and Test sets**

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

**Step 6: Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()

#Fit scalar only on training data
#Use the same scaling on test data
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```
