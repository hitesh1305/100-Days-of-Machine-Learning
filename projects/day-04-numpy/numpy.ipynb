**Array Operations**


```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print("Addition: ", a+b)
print("Subtraction: ",a-b)
print("Multiplication: ", a*b)
print("Division: ", a/b)
print("Dot product: ", np.dot(a,b))
print("Cross product: ", np.cross(a,b))
```

**Reshaping and Flatenning**

```python
arr = np.arange(1,10)
reshaped = arr.reshape(3,3)
flattened = reshaped.flatten()
print(reshaped)
print(flattened)
```

```python
arr = np.array([1,2,1,2,1])
print(np.sum(arr == 1))
```

**Broadcasting** - Vectorization

**Statistics**

```python
arr = np.array([10,20,30,40,50])

print(np.mean(arr))
print(np.sum(arr))
print(np.max(arr))
print(np.min(arr))
```

**Boolean Masking** - filtering the data

```python
arr = np.array([10,20,30,40,50])

print(arr[arr > 25])
```

```python
arr = np.array([10,20,30,40,50])

mean = np.mean(arr)
above_mean = arr[arr > mean]

print("Mean:", mean)
print("Above mean:", above_mean)
```
