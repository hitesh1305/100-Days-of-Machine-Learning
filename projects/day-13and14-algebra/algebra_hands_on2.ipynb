**DOT PRODUCTS**

```python
v1 = np.array([1,3])
v2 = np.array([4,2])
```

```python
dot = np.dot(v1,v2)
print(dot)
```

**Euclidean Distance**

```python
diff = v1-v2
print(diff)
```

linalg = linear algebra

norm = euclidean distance

```python
distance = np.linalg.norm(diff)
print(distance)
```

**Matrix Multiplication and Linear Model**

```python
X = np.array([
    [1.0, 30.0],
    [2.0, 50.0],
    [3.0, 70.0]
])

print("X:\n", X)
print("X shape: ",X.shape)
```

```python
W = np.array([
    [0.5],
    [0.1]
])

print("W:\n", W)
print("W shape: ",W.shape)
```

  X       W

(3,2) * (2,1)
   
   2 and 2

These two should be same

and result will be of shape (3,1)

```python
xw = X @ W  #np.dot() = @ 
#both do matrix multiplication just @ is easy 
print(xw)
print(xw.shape)
```

**y = XW + b**

```python
b = 2.0

y = xw +b
print(y)
print(y.shape)
```

**Connecting this to a samples dataset**

```python
X_features = data[:,:-1] #all rows all columns (except last)
X_target = data[:,-1] #all rows but only last column


print(X_features)
print(X_features.shape)


print("\n\n",X_target)
print(X_target.shape)
```
